```python
mode_im = ['date_added','rating','duration']
for i in mode_im:
    ab[i] = ab[i].fillna(ab[i].mode()[0])
```

```python
ab.drop(['show_id','date_added','description'],axis='columns', inplace= True)
ab.isnull().sum()
```

```python
ab.head(20)
```

```python
plt.figure(figsize=(6,6))
colors = ["tab:red","tab:blue"]
sns.countplot(x ='type', data = ab ,palette=colors )
plt.title("No Of Movies vs TV Shows")
plt.show()
```

```python
plt.figure(figsize=(10,10))
y = sns.countplot(y='country',
            data=ab, order =ab['country'].value_counts().index[:15])

plt.tick_params(labelsize=10)
plt.title("Country With Most Released")
plt.show()
```

```python
# analyze the distribution of shows/movies in different countries using bar plots
country_counts = ab['country'].value_counts()

plt.figure(figsize=(12, 6))
country_counts[:10].plot(kind='bar', color='blue')
plt.title('Top 10 Movie/Show Counts per Country on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Shows/Movies')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
release_year_counts = ab['release_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.scatter(release_year_counts.index, release_year_counts.values, marker='o')
plt.title('Number of Shows/Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows/Movies')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# analyze the distribution of different genres using bar plots
genre_counts = ab['listed_in'].value_counts()

plt.figure(figsize=(12, 6))
genre_counts[:10].plot(kind='bar', color='purple')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genres')
plt.ylabel('Number of Shows/Movies')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
labels = ["Movies", "TV Show"]
size = ab["type"].value_counts()
colors = plt.cm.Wistia(np.linspace(0,1,2))
explode= [0, 0.2]
plt.rcParams["figure.figsize"] = (5,5)
plt.pie(size,labels = labels, colors= colors, explode = explode, shadow= True, startangle=15,autopct="%1.2f")
plt.title("Distribution of Movies and TV Shows", fontsize=15)
plt.legend()
plt.show()
```
